```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv('preprosessing_ver8_r4.csv')

# Cumulative_Audience 열의 데이터 시각화
plt.figure(figsize=(10, 6))
plt.plot(df['Cumulative_Audience'], marker='o', linestyle='-', color='skyblue')
plt.title('Cumulative Audience')
plt.xlabel('Index')
plt.ylabel('Cumulative Audience')
plt.grid(True)
plt.show()
```
