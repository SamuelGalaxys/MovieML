```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime, timedelta
import time

# CSV 파일 읽기
df = pd.read_csv('output_to_trend.csv')

# 결과를 저장할 데이터프레임 초기화
all_trend_data = pd.DataFrame()

# 각 영화에 대해 Google 트렌드 데이터 수집
for index, row in df.iterrows():
    movie_title = row['Movie_Title']
    release_date = datetime.strptime(row['Release_Date'], '%Y-%m-%d')
    # 개봉일 전후 1주일 기간 설정
    end_date = (release_date + timedelta(days=6)).strftime('%Y-%m-%d')
    start_date = release_date.strftime('%Y-%m-%d')

    # Google 트렌드 URL 설정
    url = f'https://trends.google.com/trends/explore?date={start_date}%20{end_date}&geo=KR&q={movie_title}&hl=ko"'
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}
    
    # HTTP 요청 보내기
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # 데이터 추출 예시 (페이지 구조에 따라 조정 필요)
        # 여기서는 가상의 데이터 구조를 가정
        trends = []
        for trend_item in soup.select('div.trend-item'):  # 실제 Google 트렌드 페이지의 구조에 맞게 조정 필요
            date = trend_item.select_one('span.date').get_text(strip=True)
            value = trend_item.select_one('span.value').get_text(strip=True)
            trends.append({
                'Date': date,
                'Value': value
            })
        
        trend_data = pd.DataFrame(trends)
        trend_data['Movie_Title'] = movie_title

        # 결과 데이터프레임에 추가
        all_trend_data = pd.concat([all_trend_data, trend_data])
        
        # 요청 간 대기 시간 설정
        time.sleep(60)
    else:
        print(f"Failed to retrieve data for {movie_title}")

# 결과 데이터프레임을 CSV로 저장
all_trend_data.to_csv('trend_data.csv', index=False)
```
