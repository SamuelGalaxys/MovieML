```python
from pyspark.sql.functions import to_date

# CSV 파일 경로
csv_file_path = "0530_dataset\output_2017-01-01_2018-05-25_update.csv"

# CSV 파일을 Spark DataFrame으로 읽어오기
df = spark.read \
    .format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("treatEmptyValuesAsNulls", "true") \
    .load(csv_file_path)

# Release_Date 열을 날짜 형식으로 변환
df = df.withColumn("Release_Date", to_date(df["Release_Date"]))

# 임시 뷰 생성
df.createOrReplaceTempView("csv_data")

# 변환된 데이터를 사용하여 쿼리 실행
result = spark.sql("SELECT * FROM csv_data WHERE Release_Date <= '2018-05-25' AND Release_Date >= '2017-01-01'")

# 결과를 Pandas DataFrame으로 변환
pdf = result.toPandas()

# 결과를 CSV 파일로 저장
csv_filename = "0530_dataset\output_2017-01-01_2018-05-25_update_final.csv"
pdf.to_csv(csv_filename, index=False, encoding='utf-8-sig')
```

```python
import pandas as pd

# CSV 파일 읽기
merged_df = pd.read_csv('0530_dataset\merged_dataset_0605.csv')
preprocessing_df = pd.read_csv('0530_dataset\preprosessing_ver7_r11.csv')

# Movie_Title을 기준으로 병합 (inner join)
merged = preprocessing_df.merge(merged_df[['Movie_Title', 'Cumulative_Audience']], on='Movie_Title', how='left', suffixes=('', '_new'))

# Cumulative_Audience 값을 업데이트
merged['Cumulative_Audience'] = merged['Cumulative_Audience_new'].combine_first(merged['Cumulative_Audience'])

# 불필요한 열 삭제
merged.drop(columns=['Cumulative_Audience_new'], inplace=True)

# 결과를 새로운 CSV 파일로 저장
merged.to_csv('updated_preprosessing_ver7_r11.csv', index=False)
```

```python
import openpyxl
import csv
from bs4 import BeautifulSoup

def html_to_csv(html_file, csv_file):
    # openpyxl을 사용하여 HTML 파일을 엑셀 파일처럼 엽니다.
    wb = openpyxl.load_workbook(html_file)
    # 첫 번째 시트를 선택합니다.
    sheet = wb.active

    # CSV 파일을 쓰기 모드로 엽니다.
    with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)

        # 시트의 각 행을 반복하며 CSV 파일에 쓰기합니다.
        for row in sheet.iter_rows(values_only=True):
            writer.writerow(row)

# HTML 파일과 CSV 파일의 경로를 지정합니다.
html_file = '0530_dataset\\2020-05-28_2022-05-27.xlsx'  # 여기에 당신의 HTML 파일 경로를 넣어주세요
csv_file = 'output.csv'  # CSV 파일 이름을 원하는 이름으로 바꿔주세요

# HTML을 CSV로 변환합니다.
html_to_csv(html_file, csv_file)
```

```python
import pandas as pd

# Excel 파일을 읽어들임
df_0501 = pd.read_csv('0530_dataset\output_2018-05-26_2020-05-27_update.csv')
df_0101 = pd.read_csv('0530_dataset\output_2017-01-01_2018-05-25.csv')

# 'Release_Date'가 2022-01-01 이후인 행들을 필터링
df_0101_filtered = df_0101[pd.to_datetime(df_0101['Release_Date']) >= pd.Timestamp(2017, 1, 1)]

# 'Movie_Title'을 기준으로 두 데이터프레임을 비교하여 일치하는 경우 'Cumulative_Audience' 데이터를 업데이트
for index, row in df_0501.iterrows():
    movie_title = row['Movie_Title']
    if movie_title in df_0101_filtered['Movie_Title'].values:
        print(movie_title)
        # 일치하는 'Movie_Title'의 'Cumulative_Audience' 값을 05_28 파일에서 가져와서 01_01 파일에 업데이트
        cumulative_audience = row['Cumulative_Audience']
        df_0101.loc[df_0101['Movie_Title'] == movie_title, 'Cumulative_Audience'] = cumulative_audience

# 변경된 내용을 새 Excel 파일로 저장
df_0101.to_csv('0530_dataset\output_2017-01-01_2018-05-25_update.csv', index=False, encoding='utf-8')
```

```python
from pyspark.sql import SparkSession

# SparkSession 생성
spark = SparkSession.builder \
    .appName("HTML to CSV") \
    .getOrCreate()

# HTML 파일을 Spark DataFrame으로 읽기
html_df = spark.read.option("inferSchema", "true").html("0530_dataset\\2020-05-28_2022-05-27.xlsx")

# DataFrame의 스키마 출력
html_df.printSchema()

# 테이블로 등록하기
html_df.createOrReplaceTempView("html_table")

# SQL 쿼리를 사용하여 필요한 데이터를 선택합니다.
result_df = spark.sql("SELECT * FROM html_table")

# CSV 파일로 저장
result_df.write.csv("저장할파일명.csv", header=True)

# SparkSession 종료
spark.stop()
```

```python
import pandas as pd

# CSV 파일들의 파일 이름 리스트
file_names = ['0530_dataset\output_2022-05-28_2024-05-01_update_final.csv', '0530_dataset\output_2020-05-28_2022-05-27_update_final.csv', '0530_dataset\output_2018-05-26_2020-05-27_update_final.csv', '0530_dataset\output_2017-01-01_2018-05-25_update_final.csv']

# 빈 리스트를 만들어 데이터프레임들을 담을 준비
dfs = []

# 각 CSV 파일을 순회하면서 데이터프레임 생성 후 리스트에 추가
for file_name in file_names:
    df = pd.read_csv(file_name)
    dfs.append(df)

# 데이터프레임들을 수직으로 결합하기
result = pd.concat(dfs, axis=0)

# 결합된 데이터프레임을 새로운 CSV 파일로 저장하기
result.to_csv('0530_dataset\\5_years_moive_data.csv', index=False)
```
