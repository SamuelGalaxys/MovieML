```python
# -*- coding: utf-8 -*-
import os
import sys
import urllib.request
import json
import pandas as pd
from datetime import datetime, timedelta

client_id = "zpxnQpa1UTcH7rTjlSJy"
client_secret = "UvLNC2Ow_i"
url = "https://openapi.naver.com/v1/datalab/search"

df = pd.read_csv('0530_dataset\cleaned_preprosessing_ver3.csv')

# Dictionary to store data for each period
data_dict = {
    'week_1': {'weekdays': [], 'weekends': []},
    'week_2': {'weekdays': [], 'weekends': []},
    'week_3': {'weekdays': [], 'weekends': []}
}

for index, row in df.iterrows():
    movie_title = row['Movie_Title']
    release_date = datetime.strptime(row['Release_Date'], '%Y-%m-%d')
    
    # Periods to check: first week, second week, third week
    periods = [(0, 6), (0, 13), (0, 20)]
    
    for i, (start_offset, end_offset) in enumerate(periods):
        start_date = (release_date + timedelta(days=start_offset)).strftime('%Y-%m-%d')
        end_date = (release_date + timedelta(days=end_offset)).strftime('%Y-%m-%d')
        
        # JSON 형식의 본문 문자열을 수정
        body = f"""
        {{
            "startDate": "{start_date}",
            "endDate": "{end_date}",
            "timeUnit": "date",
            "keywordGroups": [
                {{
                    "groupName": "{movie_title}",
                    "keywords": ["{movie_title}"]
                }}
            ]
        }}
        """
    
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id", client_id)
        request.add_header("X-Naver-Client-Secret", client_secret)
        request.add_header("Content-Type", "application/json")
        
        try:
            response = urllib.request.urlopen(request, data=body.encode("utf-8"))
            rescode = response.getcode()
            if rescode == 200:
                response_body = response.read()
                response_data = json.loads(response_body.decode('utf-8'))
                
                # 결과 리스트에서 데이터 추출
                results = response_data['results'][0]
                data = results['data']
                
                # Convert to DataFrame
                df_temp = pd.DataFrame(data)
                df_temp['period'] = pd.to_datetime(df_temp['period'])
                
                # Separate weekdays and weekends and append to respective lists
                for index, row in df_temp.iterrows():
                    if row['period'].dayofweek < 5:
                        data_dict[f'week_{i+1}']['weekdays'].append(row['ratio'])
                    else:
                        data_dict[f'week_{i+1}']['weekends'].append(row['ratio'])
                
            else:
                print("Error Code:" + str(rescode))
        except Exception as e:
            print(f"Error: {e}")

# Calculate statistics for each period
stats = {}

for period, data in data_dict.items():
    weekdays_series = pd.Series(data['weekdays'])
    weekends_series = pd.Series(data['weekends'])
    
    stats[period] = {
        'weekdays_mean': weekdays_series.mean(),
        'weekdays_std': weekdays_series.std(),
        'weekends_mean': weekends_series.mean(),
        'weekends_std': weekends_series.std()
    }

# 결과 출력
for period, stat in stats.items():
    print(f"{period} 데이터에 대한 통계:")
    print(f"평일 평균: {stat['weekdays_mean']}")
    print(f"평일 표준편차: {stat['weekdays_std']}")
    print(f"주말 평균: {stat['weekends_mean']}")
    print(f"주말 표준편차: {stat['weekends_std']}")
    print()
```
